```python
# subplots using add_subplot, because you can't create blank figure
import matplotlib.pyplot as plt
import pandas as pd
#Plots in matplotlib reside within a figure object, use plt.figure to create new figure
fig=plt.figure()
#Create one or more subplots using add_subplot, because you can't create blank figure
ax = fig.add_subplot(1,1,1)
#Variable
ax.hist(df['EmpID'],bins = 3000)
#Labels and Tit
plt.xlabel('LocationCode')
plt.ylabel('Current Employee Rating')
plt.show()
```

```python
#Create one or more subplots using add_subplot, because you can't create blank figure
ax = fig.add_subplot(1,1,1)
#Variable
ax.scatter(df['EmpID'],df['DepartmentType'])
#Labels and Tit
plt.xlabel('EmployeeType')
plt.ylabel('JobFunctionDescription')
plt.show()
```
